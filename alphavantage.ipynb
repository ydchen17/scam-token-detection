```python
import requests
import pandas as pd
import json
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```python
api_key="T6VX725RGZZJFF8P"
url = 'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=LUNA&market=USD&apikey={api_key}'
r = requests.get(url)
response_dict = r.json()
_, header = r.json()

df = pd.DataFrame.from_dict(response_dict[header], orient='index')
df_cols = [i.split(' ')[1] for i in df.columns]
df.columns = df_cols
df = df.astype('float')
df.index = pd.to_datetime(df.index)
df = df.loc[:,~df.columns.duplicated()]
#df["2021-01-25":].close.plot(title="Price of Terra (LUNA)", figsize = (10, 4))
df.close.plot(title="Price of Terra (LUNA)", figsize = (10, 4))
```
